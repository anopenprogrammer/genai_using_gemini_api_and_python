**Chat Conversation**

```python
import google.generativeai as genai
import pathlib
import textwrap
from IPython.display import display
from IPython.display import Markdown
```

```python
def to_markdown(text):
  text = text.replace('•',' *')
  return Markdown(textwrap.indent(text, '>'))
```

```python
#Used to store securely your API key
from google.colab import userdata
```

```python
GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')
```

```python
genai.configure(api_key=GOOGLE_API_KEY)
```

```python
model3 = genai.GenerativeModel('gemini-1.0-pro-latest')
```

```python
model3
```

```python
model3.count_tokens("What is meaning of life?")
```

```python
chat = model3.start_chat(history=[])
```

```python
response = chat.send_message("What is meaning of life? I want to explain to 5 years old")
```

```python
response.text
```

```python
chat.history
```

```python
for message in chat.history:
  display(to_markdown(f'**{message.role}**: {message.parts[0].text}'))
```

**Use Embedding**

```python
result = genai.embed_content(
    model="models/embedding-001",
    content=["The food was delicious and the waiter...",
             "Why this is happening?"
             ],
    task_type="retrieval_document",
    title="Embedding of single string"
)
```

```python
len(result["embedding"])
```

```python
result["embedding"]
```

```python
for i in result["embedding"]:
  print(len(i))
```
